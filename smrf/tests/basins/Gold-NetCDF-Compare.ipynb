```python
import xarray as xr
import matplotlib.pyplot as plt
import math
```

# Description

Notebook to compare new and current gold files.

This notebook creates spatial difference plots between two NetCDF files, subtracting the "new" from the "old". This comparison should be performed whenever there is a need to update the "gold" references files for the tests.

This notebook expects the "new" file to be in the directory, where this notebook is and relative to the test basin folders.

```python
%cd /data/projects/iSnobal-CM/smrf/smrf/tests/basins/
```

```python
filename = "net_solar_hrrr_albedo"
variable = "net_solar"
```

```python
old_nc = xr.open_dataset(f"Lakes/gold_hrrr/{filename}.nc")
new_nc = xr.open_dataset(f"Lakes/output/{filename}.nc")
```

```python
diff = (new_nc[variable] - old_nc[variable])
```

```python
plots = new_nc[variable].plot(x='x', y='y', col="time", figsize=(11, 5))
plots.map(lambda: plt.gca().set_aspect("equal"));
```

```python
plots = old_nc[variable].plot(x='x', y='y', col="time", figsize=(11, 5))
plots.map(lambda: plt.gca().set_aspect("equal"));
```

```python
plots = diff.plot(x='x', y='y', col="time", figsize=(11, 5))
plots.map(lambda: plt.gca().set_aspect("equal"));
# plots.axes.set_aspect("equal");
```

```python
old_nc.close()
new_nc.close()
```

```python
plots.fig.savefig(f"{filename}_{variable}_xy.png", dpi=120)
```

```python
num_times = diff.sizes["time"]
num_rows = math.ceil(num_times / 2)
```

```python
fig, axes = plt.subplots(num_rows, 2, figsize=(10, 4 * num_rows), dpi=120)
axes = axes.flatten()

for i in range(num_times):
    diff.isel(time=i).plot.hist(ax=axes[i], bins=30, color='skyblue')
```

```python
fig.savefig(f"{filename}_{variable}_hist.png", dpi=120)
```
