# Description

Notebook to compare topo.nc file differences.

Use this to compare difference to any variable in the file.

## Set path to the local test `basins` directory.
Example: `/data/iSnobal/smrf/smrf/tests/basins/`

```python
%cd /data/projects/iSnobal-CM/smrf/smrf/tests/basins/
```

```python
filename = "topo"
variable = "sky_view_factor"
```

### Relative path to the `topo.nc` based on the above local directory

```python
old_nc = xr.open_dataset(f"Lakes/topo/{filename}.nc")
new_nc = xr.open_dataset(f"{filename}.nc")
```

```python
diff = (new_nc[variable] - old_nc[variable])
```

### Current file

```python
plot = old_nc[variable].plot(x='x', y='y', figsize=(5, 5))
plot.axes.set_aspect("equal")
```

### New file

```python
plot = new_nc[variable].plot(x='x', y='y', figsize=(5, 5))
plot.axes.set_aspect("equal")
```

```python
plot = diff.plot(x='x', y='y', figsize=(5, 5))
plot.axes.set_aspect("equal");
```

```python
old_nc.close()
new_nc.close()
```

```python
plot.figure.savefig(f"{filename}_{variable}_xy.png", dpi=120)
```

```python
fig = plt.figure(figsize=(5, 5))
diff.plot.hist(bins=30, color='skyblue', ax=fig.gca());
```

```python
fig.savefig(f"{filename}_{variable}_hist.png", dpi=120)
```
